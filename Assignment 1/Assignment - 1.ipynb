```python
# defining arms
arm1 = (1,3)
arm2 = (-3,8)
arm3 = (2,5)
arm4 = (-2,6)
arm5 = (3,4)
arm6 = (-2,2)

arms = [arm1,arm2,arm3,arm4,arm5,arm6]
```

### Given is a six-armed bandit,
- The first arm shall sample its reward uniformly from the interval [1, 3).
- The second arm shall sample its reward uniformly from [-3, 8).
- The third arm shall sample its reward uniformly from the interval [2, 5).
- The fourth arm shall sample its reward uniformly from [–2, 6).
- The fifth arm shall sample its reward uniformly from [3, 4).
- The sixth arm shall sample its reward uniformly from [-2, 2).

#### 1.1 What is the expected reward when actions are chosen uniformly?

```python
# 1.1
sum_=0
for item in arms:
    sum_ += (item[0] + item[1])/2
    
sum_/6
```

#### 1.2 Implement the six-armed bandit from 1.1) and compute the sample average reward for 10 uniformly chosen actions.
#### Compare this to your expectation from 1.1).

```python
# 1.2

iters = 10

sum_ = 0
for i in range(iters):
    arm = arms[random.randint(0,len(arms) - 1)]
    sum_ += np.random.uniform(arm[0],arm[1])
    
sum_/iters
    
```

#### 1.3) Initialize Q(a i )=0 and chose 4000 actions according to an ε-greedy selection strategy (ε=0.1).
#### Update your action values by computing the sample average reward of each action recursively.
#### For every 100 actions show the percentage of choosing arm 1, arm 2, arm 3, arm 4, arm 5, and
#### arm 6 as well as the resulting average reward.

```python
# 1.3
iters_rl = 4000
epsilon = 0.1
#initialising Q and N
Q = np.zeros((len(arms)))
N = np.zeros((len(arms)))

#initialising the reward model
arms = [arm1,arm2,arm3,arm4,arm5,arm6]


for i in range(1,iters_rl+1):
    a = np.argmax(Q)
    prob = np.random.uniform(0,1)
    if prob <= epsilon:
        a = random.randint(0,len(arms) - 1)
    
    if i%100 == 0:
        print("Iteration ",i)
        print("Average Reward %.2f" % (Q.mean()))
        print("Percentage of choosing each arm")
        per = N/N.sum()
        
        k=1
        for item in per:
            print("For Arm%d %.0f percent" % (k,item*100) )
            k += 1
        print("---------------------")
        
    
    
        
    arm = arms[a]
    r = np.random.uniform(arm[0],arm[1])
    N[a] = N[a] + 1
    Q[a] = Q[a] + (r-Q[a])/N[a]
```
